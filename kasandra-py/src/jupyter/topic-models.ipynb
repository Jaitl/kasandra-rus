# Тематическое моделирование

[gensim](https://radimrehurek.com/gensim/wiki.html)

```python
import logging, gensim, bz2
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
dictionary = gensim.corpora.Dictionary.load('/data/kasandra/gensim/news-4965.dict')
corpus = gensim.corpora.MmCorpus('/data/kasandra/gensim/news-4965.mm')

print(dictionary)
print(corpus)
```

## LSA

```python
lsi = gensim.models.lsimodel.LsiModel(corpus=corpus, id2word=dictionary, num_topics=100)
```

```python
lsi.print_topics(10)
```

## LDA

```python
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=100, update_every=1, chunksize=10000, passes=1)
```

```python
lda.print_topics(20)
```
