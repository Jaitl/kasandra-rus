```python
import json
from collections import Counter
```

```python
class News:
    def __init__(self, id, date, title, content, url, siteType):
        self.id = id
        self.date = date
        self.title = title
        self.content = content
        self.url = url
        self.siteType = siteType
    
    @classmethod
    def from_json(cls, json_str):
        json_dict = json.loads(json_str)
        return cls(**json_dict)
```

```python
news = []
with open('/data/kasandra/year/10k.test.normalized.json', encoding="utf8") as f:
    for line in f:
        news.append(News.from_json(line))
```

```python
counts = []
all_tokens = []
for n in news:
    tokens = n.content.split(" ")
    if len(tokens) >= 100:
        all_tokens.extend(tokens)
        counts.append(len(tokens))
        
print("Max: %s" % max(counts))
print("Min: %s" % min(counts))
```

```python
counts = Counter(all_tokens)
one_time = [k for k, v in dict(counts).items() if v == 1]

print("Всего токенов: %s" % len(all_tokens))
print("Значимых токенов: %s" % (len(all_tokens) - len(one_time)))
```

```python
n_gram_count = 2
```

```python
count_tokens = len(all_tokens)
ngramm_list = []
i = 0
while i < count_tokens:
    n_arr = []
    
    if i + n_gram_count <= count_tokens:
        k = 0
        while k < n_gram_count:
            n_arr.append(all_tokens[i+k])
            k = k + 1
    else:
        break;
    i = i + 1
            
    ngramm_list.append(" ".join(n_arr))   
```

```python
counter = Counter(ngramm_list)
counter.most_common()
```

```python
frequent_ngramm = [k for k, v in dict(counter).items() if v < 50]
print("Количество частых словосочетаний: %s" % len(frequent_ngramm))
```
