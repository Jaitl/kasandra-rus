# Входной текст

```python
text = """Pokemon Go попал в руки общественности только в июле этого года, но уже было достаточное количество обновлений с момента его запуска. Игра была встречена в основном положительными отзывами, когда она была впервые выпущена, но после первоначального ажиотажа некоторые критические отзывы о том, что игра слишком легкая, стали снижать к ней интерес. Разработчики Niantic приняли критику близко к сердцу, и в результате им удалось добиться того, чтобы выпустить частые обновления к игре, которые добавляют все больше и больше возможностей для пользователей, сообщает ITbukva.com.
В последнем обновлении версии 0.37.0 для Android и 1.7.0 для iOS, Niantic, наконец, представят новую функцию Buddy, которую они официально объявили еще 2 сентября."""
print(text)
```

# [Yandex MyStem](https://tech.yandex.ru/mystem/doc/index-docpage/)

Производит токенизацию и лемминг. Приводит слова к их начальному состоянию.

```python
from pymystem3 import Mystem
mystem = Mystem(entire_input=False)

mystem.lemmatize(text)
```

## Морфологический разбор

```python
mystem.analyze(text)
```

# [NLTK 3.0](http://www.nltk.org/)

Токенизация производится отдельно, есть несколько токенизаторов. После токенизации производится стеминг - обрезаются приставки, окончания и т.д, остаются только корни

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.snowball import SnowballStemmer
tokenizer = RegexpTokenizer('[А-ЯЁа-яёA-Za-z]+')
tokens = tokenizer.tokenize(text)
stemmer = SnowballStemmer("russian")
stemTokens = list(map(lambda word: stemmer.stem(word), tokens))
stemTokens
```

# [pymorphy2](https://pymorphy2.readthedocs.io/en/latest/user/guide.html)

Морфологический анализатор, умеет:
* приводить слово к нормальной форме (например, “люди -> человек”, или “гулял -> гулять”).
* ставить слово в нужную форму. Например, ставить слово во множественное число, менять падеж слова и т.д.
* возвращать грамматическую информацию о слове (число, род, падеж, часть речи и т.д.)

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
morphTokens = list(map(lambda word: morph.parse(word)[0].normal_form, tokens))
morphTokens
```

## Морфологический разбор

```python
list(map(lambda word: morph.parse(word), tokens))
```
